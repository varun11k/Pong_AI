```python
import sys, pygame, random, time

#Initial setup
pygame.init()
clock = pygame.time.Clock()

#Creating the field for pong
screen_width = 800
screen_height = 400
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('PONG')

#Creating the elements to be used in the game
pong = pygame.Rect(screen_width/2-10,screen_height/2-10,20,20)
HU = pygame.Rect(screen_width-20,screen_height/2-20,10,40)
AI = pygame.Rect(10,screen_height/2-20,10,40)
#res = pygame.Rect(screen_width/2-10,screen_height/2+20, 20, 40)

#Creating the pugments used in the screen
bg_col = pygame.Color('grey0')
elmt_col=(245,245,245)
ai_col = (255,0,0)
hu_col = (0,0,255)
aaline_col = (255,255,255)
#white = (255,255,255)
#res_col = (0,255,0)
#go_col = (0,0,0)

#Setting the speed for the pong
pong_speed_x = 6 * random.choice((1,-1))
pong_speed_y = 6 * random.choice((1,-1))
HU_sp = 0
AI_sp = 5

#Text variables
YOU_score = 0
AI_score = 0
score_font = pygame.font.Font("freesansbold.ttf",30)


#time variable
score_time = None

#Sound variables
pong_sound_wall = pygame.mixer.Sound("wall_bounce.ogg")
pong_sound_pad = pygame.mixer.Sound("pad_bounce.ogg")
pong_sound_score = pygame.mixer.Sound("point.ogg")


def pong_animation():
    global pong_speed_x, pong_speed_y, YOU_score, AI_score, score_time
    #Creating the moving animation
    pong.x += pong_speed_x
    pong.y += pong_speed_y
    
    
    #Setting the boundry for the pong
    if pong.top <= 0 or pong.bottom >= screen_height :
        pygame.mixer.Sound.play(pong_sound_wall)
        pong_speed_y = pong_speed_y * -1
        
    #Adding the codes to increase score count if human scores    
    if pong.left <= 0 and YOU_score < 10:
        pygame.mixer.Sound.play(pong_sound_score)
        YOU_score += 1
        score_time = pygame.time.get_ticks()
    if YOU_score == 10:
        YOU_score = 0
        AI_score = 0
        pong_restart()
    
    #Adding the codes to increase score count if AI scores        
    if pong.right >= screen_width and AI_score < 10:
        pygame.mixer.Sound.play(pong_sound_score)
        AI_score += 1
        score_time = pygame.time.get_ticks()
    if AI_score == 10:
        AI_score = 0
        YOU_score = 0
        pong_restart()
       
        
        
        
        
    #implementing collide and rebound
    
    if pong.colliderect(HU) and pong_speed_x >0:
        pygame.mixer.Sound.play(pong_sound_pad)
        if abs(pong.right - HU.left) < 10:
            pong_speed_x *= -1
        elif abs(pong.bottom - HU.top) < 10 and pong_speed_y > 0 :
            pong_speed_y *= -1
        elif abs(pong.top - HU.bottom) < 10 and pong_speed_y < 0 :
            pong_speed_y *= -1    
    
    
    if pong.colliderect(AI) and pong_speed_x <0:
        pygame.mixer.Sound.play(pong_sound_pad)
        if abs(pong.left - AI.right) < 10:
            pong_speed_x *= -1
        elif abs(pong.bottom - AI.top) < 10 and pong_speed_y > 0 :
            pong_speed_y *= -1
        elif abs(pong.top - AI.bottom) < 10 and pong_speed_y < 0 :
            pong_speed_y *= -1    


#def restart_game():
#    mouse = pygame.mouse.get_pos()
#    click = pygame.mouse.get_pressed()
#    screen.fill(white)
#    if ((screen_width/2-10) > mouse[0] > (screen_width/2+10)) and ((screen_height/2+20)> mouse[1] > (screen_height/2-20)):
#        pygame.draw.rect(screen,res_col,res)
#        go = score_font.render("GO!",False,go_col)
#        screen.blit(go,(screen_width/2+5,screen_height/2+10))        
#        if click[0] == 1:
#            intro()
            
            
#def you_win_func():            
#    win_you = score_font.render("YOU WIN",False,hu_col)
#    screen.blit(win_you,(screen_width/2+50,100))
#    event1 = True
#    while event1:
#        for event in pygame.event.get():
#            if event.type == pygame.KEYDOWN:
#                if event.key == pygame.K_RETURN:
#                    pong_restart()
                    
#def ai_win_func():
#    win_ai = score_font.render("AI WINS",False,ai_col)
#    screen.blit(win_ai,(screen_width/2-50,100))
#    event2 = True
#    while event2:
#        for event in pygame.event.get():
#            if event.type == pygame.KEYDOWN:
#                if event.key == pygame.K_RETURN:
#                    pong_restart()                    
    

def human_animation():
    HU.y += HU_sp
    
    #Setting boundry for Human
    if HU.top <= 0:
        HU.top = 0
    if HU.bottom >= screen_height:
        HU.bottom = screen_height
        
def ai_animation():
    if pong.x < screen_width/2:
        if AI.top < pong.y:
            AI.top += AI_sp
        if AI.bottom > pong.y:
            AI.bottom -= AI_sp
        if AI.top <=0:
            AI.top = 0
        if AI.bottom >= screen_height:
            AI.bottom = screen_height    
        
        
def pong_restart():
    global pong_speed_x, pong_speed_y, score_time
    current_time = pygame.time.get_ticks()
    pong.center = (screen_width/2, screen_height/2)
    
    if current_time - score_time < 700:
        num_3 = score_font.render("3",False,elmt_col)
        screen.blit(num_3,(screen_width/2+20,20))
    if 700 < current_time - score_time < 1400:
        num_2 = score_font.render("2",False,elmt_col)
        screen.blit(num_2,(screen_width/2+20,20))
    if 1400 < current_time - score_time < 2100:
        num_1 = score_font.render("1",False,elmt_col)
        screen.blit(num_1,(screen_width/2+20,20))    
    
    
    if current_time - score_time < 2100:  
        pong_speed_x, pong_speed_y = 0,0
    else:
        pong_speed_x = 5 * random.choice((1,-1))
        pong_speed_y = 5 * random.choice((1,-1))
        score_time = None

        

    
game_loop = True        


while game_loop:
    #handling the input
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
         #Player movement
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_DOWN:
                HU_sp += 5
            if event.key == pygame.K_UP:
                HU_sp -= 5
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_DOWN:
                HU_sp -= 5
            if event.key == pygame.K_UP:
                HU_sp += 5
    
 

    
    #drawing the elements onto the screen
    screen.fill(bg_col)
    pygame.draw.rect(screen,hu_col,HU)
    pygame.draw.rect(screen,ai_col,AI)
    pygame.draw.ellipse(screen,elmt_col,pong)
    pygame.draw.aaline(screen,aaline_col,(screen_width/2,0),(screen_width/2,screen_height))
    
   
    
    
        
    
    #calling animation function
    pong_animation()
    human_animation()
    ai_animation()
    
    
    if score_time:
        pong_restart()
    
    hu_text = score_font.render(f"{YOU_score}",False,hu_col)
    screen.blit(hu_text,(410,370))
    
    ai_text = score_font.render(f"{AI_score}",False,ai_col)
    screen.blit(ai_text,(380,370))
    
    pygame.display.flip()
    clock.tick(60)
```
